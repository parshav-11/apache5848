```python
import cv2
import os
import argparse
import numpy as np
import PIL
from PIL import Image
from __future__ import print_function
import time
```

```python


#Initialize Counter for Bag
count = 0
x = []
#create Background Subtractor objects
backSub = cv2.createBackgroundSubtractorMOG2()    
#Initialize Camera Object
cap = cv2.VideoCapture(r"D:\Images\vid_9.mp4")
while cap.isOpened():
    #Read Video Frame by Frame
    ret, frame = cap.read()
    
    #resize frame change coordinates
    new_size = (500,300)
    frame = cv2.resize(frame, new_size)
#     frame = frame[110:300,40:250]
    
    #Cut Frame to Desired Dimensions
#     frame = frame[300:1000,100:1500]

    #update the background model
    fgMask = backSub.apply(frame)
    
    cv2.rectangle(frame, (10, 2), (100,20), (255,0,0), 1)
    cv2.putText(frame, str(cap.get(cv2.CAP_PROP_POS_FRAMES)), (15, 15),
               cv2.FONT_HERSHEY_SIMPLEX, 0.5 , (0,0,0))
    
    #Convert to Gray Scale First
    grayscale = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    #Now Convert Gray Frame Using Threshold Values to Black and White Frame
    ret, threshold = cv2.threshold(grayscale, 135, 255, cv2.THRESH_BINARY)
    
    # Detect the pattern
    contours, heirarchy = cv2.findContours(threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_TC89_KCOS)
    for cnt in contours:
        (x, y, w, h) = cv2.boundingRect(cnt)
#         list.append(cnt)
        
        # Find Area of the Object Passed
        area = cv2.contourArea(cnt)
        cv2.putText(threshold, str(area), (x, y), 0, 1, (255, 255, 255))
        cv2.rectangle(threshold, (x, y), (x + w, y + h), (255, 0, 0), 1)
        
        #area according to object area passing through frame,
        if 18100 < area:
            print(area)
#             print(len(x))
            count += 1
            print(count)
        
#     #Display Frames
#     cv2.imshow('frame', frame)
#     cv2.imshow('grayscale', grayscale)
    cv2.imshow('threshold', threshold)
    
    #Break Control
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

#Exit Gracefully
cap.release()
cv2.destroyAllWindows()
```
